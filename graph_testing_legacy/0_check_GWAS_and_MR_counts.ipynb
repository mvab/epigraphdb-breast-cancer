```python
for i in df['exposure.id']:
    if "eqtl" not in i:
        print(i)
```

```python
mreve_exp=pd.DataFrame()

for GWAS_ID in df['exposure.id']:
     if "eqtl" not in GWAS_ID:

        query="""
            MATCH (exposure:Gwas)-[mr:MR_EVE_MR]->(outcome:Gwas)
            WHERE exposure.id = '{gwas_id}' 
            RETURN exposure.id,  outcome.id,  mr.pval
        """.format(gwas_id=GWAS_ID)
        
        mr = query_to_df(session, query, print_query = False)
        mreve_exp=mreve_exp.append(mr, ignore_index=True)

```

```python
mreve_out=pd.DataFrame()

for GWAS_ID in df['outcome.id']:
     if "eqtl" not in GWAS_ID:

        query="""
            MATCH (exposure:Gwas)-[mr:MR_EVE_MR]->(outcome:Gwas)
            WHERE outcome.id = '{gwas_id}' 
            RETURN exposure.id,  outcome.id,  mr.pval
        """.format(gwas_id=GWAS_ID)
        
        mr = query_to_df(session, query, print_query = False)
        mreve_out=mreve_out.append(mr, ignore_index=True)
```
